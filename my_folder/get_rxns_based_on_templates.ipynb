```python
# Import required library
import pandas as pd

# Read your dataframe (replace 'your_data.csv' with your actual file path)
df = pd.read_csv('csv_files_non_stereo/reaction_templates_validated.csv', low_memory=False, sep="\t")

# Given template (replace with your template)
template = "[NH;D2;+0:7]-[c;H0;D3;+0:5]1:[n;H0;D2;+0:4]:[c;H0;D3;+0:3]:[c;H0;D3;+0:2]:[cH;D2;+0:1]:[n;H0;D2;+0:6]:1>>C-N(-C)-[CH;D2;+0:1]=[C;H0;D3;+0:2]-[C;H0;D3;+0:3]=O.[NH2;D1;+0:4]-[C;H0;D3;+0:5](=[NH;D1;+0:6])-[NH;D2;+0:7]"

# Find matches (first try template, then hash)
matches = df[
    (df['RetroTemplate'] == template) | 
    (df['TemplateHash'] == template)
]

# Select only the columns we want
result = matches[['RxnSmilesClean', 'TemplateHash', 'RetroTemplate']]

# Export to CSV (replace 'output.csv' with your desired filename)
result.to_csv('output.csv', index=False)

# Display number of matches found
print(f"Found {len(result)} matching reactions")
print("\nFirst few matches:")
print(result.head())
```
